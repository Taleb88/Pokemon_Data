```python
# grabbing rows with attack >= 100 and defense >= 80
new_df = df.loc[(df['Attack'] >= 100) & (df['Defense'] >= 80)]
```

```python
new_df
```

```python
# saving filtered results in new csv/workbook
new_df.to_csv('filtered_data.csv', index=False)
new_df.to_excel('filtered_data.xlsx',index=False)
```

```python
# index reset
new_df.reset_index(drop=True, inplace=True)
```

```python
new_df
# saving filtered results in new csv/workbook
new_df.to_csv('filtered_data.csv', index=False)
new_df.to_excel('filtered_data.xlsx',index=False)
```

```python
for index,row in new_df.iterrows():
    print(index, row['Name'])
```

```python
# filtering out rows with names containing "Sala" or "sala"
new_df.loc[new_df['Name'].str.contains('(?i)sala')]
```

```python
# filtering out rows that do not contain "Sala" or "sala"
new_df.loc[~new_df['Name'].str.contains('(?i)sala')]
```

```python
# filtering out rows that contain "Poison" or attack <= 150
new_df.loc[new_df['Type 2'].str.contains('(?i)poison|nan', na=False)]
```

```python
new_df
```

```python
# replace "Fairy" values with "Magical"
new_df.loc[new_df['Type 2'] == 'Fairy', 'Type 2'] = 'Magical'
```
